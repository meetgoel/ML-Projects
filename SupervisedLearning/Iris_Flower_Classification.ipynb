##Iris Flower Classification

The Iris dataset is one of the most well-known datasets in the field of machine learning, frequently used for classification tasks. It consists of 150 samples of iris flowers, with four features recorded for each sample: sepal length, sepal width, petal length, and petal width. The dataset includes three different species of iris flowers: Setosa, Versicolor, and Virginica, with 50 samples for each species. This project aims to explore the Iris dataset and apply machine learning techniques, including classification algorithms, to predict the species of iris flowers based on their attributes. Through this analysis, we will gain insights into data preprocessing, feature selection, model evaluation, and overall machine learning workflows.

Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

Reading the Dataset


```python
df = pd.read_csv('/iris.csv')
df.shape
```

Splitting the Dataset

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train.shape, df_test.shape
```

Exploratory Data Analysis(EDA)

```python
df_train.head()
```

```python
df_train.info()
```

```python
df_train.describe()
```

```python
df_train.isnull().sum()
```

Data Visualization

```python
sns.pairplot(df_train, hue='target')
plt.show()
```

Data Preprocessing

```python
x_train = df_train.drop(['id','target'],axis=1)
x_test = df_train.drop(['id','target'],axis=1)
y_train = df_train['target']
y_test = df_train['target']
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape, y_test.shape)
```

Feature Engineering

Model Creation

Model-1: Logistic Regression

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

Model-2: Support Vector Machines - Classification

```python
svm = SVC()
svm.fit(x_train,y_train)
```

Model-3: Decision Tree Classifier

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

Model-4: K-Nearest Neighbors Classification

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

Model-5: Random Forest Classifier

```python
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

Model Evaluation

Logistic Regression

```python
lr_predict = lr.predict(x_test)
lr_accuracy = accuracy_score(lr_predict,y_test)
lr_accuracy*100
```

Support Vector Machines - Classification

```python
svm_predict = svm.predict(x_test)
svm_accuracy = accuracy_score(svm_predict,y_test)
svm_accuracy*100
```

Decision Tree Classifier

```python
dt_predict = dt.predict(x_test)
dt_accuracy = accuracy_score(dt_predict,y_test)
dt_accuracy*100
```

K-Nearest Neighbors Classification

```python
knn_predict = knn.predict(x_test)
knn_accuracy = accuracy_score(knn_predict,y_test)
knn_accuracy*100
```

Random Forest Classification

```python
rfc_predict = rfc.predict(x_test)
rfc_accuracy = accuracy_score(rfc_predict,y_test)
rfc_accuracy*100
```

Confusion Matrix

Logistic Regression

```python
cm = confusion_matrix(lr_predict,y_test)
sns.heatmap(cm, annot=True)
plt.show()
```

Decision Tree Classifier

```python
cm = confusion_matrix(dt_predict,y_test)
sns.heatmap(cm, annot=True)
plt.show()
```

Random Forest Classifier

```python
cm = confusion_matrix(rfc_predict,y_test)
sns.heatmap(cm, annot=True)
plt.show()
```

Classification Report

Decision Tree Classifier

```python
print(classification_report(dt_predict,y_test))
```

Random Forest Classifier

```python
print(classification_report(rfc_predict,y_test))
```

Model Optimization


Logistic Regression

```python
grid_params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 300]
}

grid_search = GridSearchCV(estimator=lr, param_grid=grid_params, cv=5)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_*100
```

Decision Tree Classifier

```python
grid_params = {
     'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random']
}

grid_search = GridSearchCV(estimator=dt, param_grid=grid_params, cv=5)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_*100
```

K-Nearest Neighbors Classifier

```python
grid_params = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [20, 30, 40]
}

grid_search = GridSearchCV(estimator=knn, param_grid=grid_params, cv=5)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_*100
```

Support Vector Machines - Classification

```python
grid_params = {
     'C': [0.001, 0.01, 0.1, 1, 10],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto'],
    'degree': [3, 4, 5],
    'class_weight': [None, 'balanced']
}

grid_search = GridSearchCV(estimator=svm, param_grid=grid_params, cv=5)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_*100
```

Random Forest Classifier

```python
grid_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

grid_search = GridSearchCV(estimator=rfc, param_grid=grid_params, cv=5, n_jobs=1)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_*100
```

Model Deployment

```python
new_data = [[4.6,3.6,1.0,0.2]]
prediction = dt.predict(new_data)
print("Predicted class: ", prediction)
```

Conclusion:

After training multiple models and performing hyperparameter tuning, it was determined that Logistic Regression is the best model for the Iris Flower Classification task. The data's linear separability makes Logistic Regression a suitable choice. By using GridSearchCV, we were able to optimize the model and achieve the highest possible score, confirming its effectiveness in this classification problem.
